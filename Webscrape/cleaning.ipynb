```python
import pandas as pd
```

```python
housing = pd.read_csv('housing.csv')
```

```python
housing.head(10)
```

```python
housing[['utca','kerület']] = housing['Address'].str.split(',', n=1, expand = True)
housing.head()
```

```python
housing['szobák'] = housing['Rooms'].map(lambda x: x.lstrip('+-').rstrip('szoba'))
housing.head()
```

```python
housing['terület'] = housing['Size'].map(lambda x: x.strip('+-').rstrip('m² terület'))
housing.head()
```

```python
housing['ár'] = housing['Price'].map(lambda x: x.strip('+-').rstrip('M Ft'))
housing.head()
```

```python
housing['terasz'] = housing['Balcony'].map(lambda x: x.strip('+-').rstrip('m² erkély'))
housing.head()
```

```python
housing['cím'] = housing['utca'].map(lambda x: x.strip('+-').rstrip('0123456789 - /A'))
housing.head(21)
```

```python
housing['kerület'] = housing['kerület'].map(lambda x: x.strip('+-').rstrip('. kerület'))
housing.head()
```

```python
housing['szobák'] = housing['szobák'].map(lambda x: x.strip('+-').rstrip('fél szoba'))
housing.head()
```

```python
housing[['szobák2','fél_szobák']] = housing['szobák'].str.split('+', n=1, expand = True)
housing
```
